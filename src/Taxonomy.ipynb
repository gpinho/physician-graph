```python
def tokenize(doc):
    '''
    INPUT: string
    OUTPUT: list of strings

    Tokenize and stem/lemmatize the document.
    '''
    return [WordNetLemmatizer().lemmatize(word) for word in word_tokenize(doc.lower())]
```

```python
vectorizer2 = TfidfVectorizer(stop_words='english', tokenizer=tokenize)
```

```python
vectors2 = vectorizer2.fit_transform(new_df_taxonomy['Definition'].fillna(' ')).toarray()
```
