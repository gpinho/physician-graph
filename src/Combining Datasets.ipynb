```python
referral_df.fillna(0.0, inplace=True)
```

```python
cols = referral_df.columns
referral_df['Average Number Unique Beneficiaries'] = 0

for col in cols:
    referral_df['Average Number Unique Beneficiaries'] += referral_df[col]

referral_df['Average Number Unique Beneficiaries'] / len(cols)

edges_df = referral_df.drop(columns=['Number Unique Beneficiaries_2009','Number Unique Beneficiaries_2010','Number Unique Beneficiaries_2011','Number Unique Beneficiaries_2012','Number Unique Beneficiaries_2013','Number Unique Beneficiaries_2014','Number Unique Beneficiaries_2015'])
edges_df.to_csv('../data/samples/physician-shared-patient-patterns-average-unique-beneficiaries-subsample.csv')
```
